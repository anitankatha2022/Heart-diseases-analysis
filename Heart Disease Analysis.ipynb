# Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pypl as plt
%matplotlib inline
```

# Loading dataset

```python
dataset=pd.read_csv('C:\\Users\\USER\\Desktop\\herdis\\heart_cleveland_upload.csv')
dataset.head(5)
```

```python
# Renaming columns
dataset.columns = ['age', 'sex', 'chest_pain_type', 'resting_blood_pressure', 'cholesterol', 'fasting_blood_sugar', 'rest_ecg', 'max_heart_rate_achieved',
       'exercise_induced_angina', 'st_depression', 'st_slope', 'num_major_vessels', 'thalassemia', 'condition']
```

```python
sns.set(style="darkgrid")
pd.crosstab(dataset.sex,dataset.condition).plot(kind="bar",figsize=(6,5),color=['#BAD720',"#33A1C9"])
plt.title('Heart Disease v/s Sex',fontsize=15)
plt.xlabel('Sex',fontsize=12)
plt.xticks(rotation=0)
plt.legend(["Do not have heart Disease", "Have heart Disease"])
plt.ylabel('Frequency',fontsize=12)
plt.show()
```

```python
sns.set(style="darkgrid")
pd.crosstab(dataset.st_slope, dataset.condition).plot(kind="bar", figsize=(8,5), color=["red", "#FFA54F"], ax=plt.gca())
plt.legend(["Do not have heart disease", "Have heart disease"])
plt.title("Heart Disease Frequency for Slope",fontsize=15)
plt.xlabel("The Slope of The Peak Exercise ST Segment",fontsize=12)
plt.xticks(rotation=0)
plt.gca().set_xticklabels(labels=["Downsloping","Flat","Upsloping"])
plt.ylabel("Frequency",fontsize=12)
plt.show()
```

```python
sns.set(style="darkgrid")
pd.crosstab(dataset.chest_pain_type, dataset.condition).plot(kind="bar", figsize=(10, 5), color=["#FF82AB", "blue"], ax=plt.gca())
plt.title("Heart Disease Frequency According To Chest Pain Experience",fontsize=15)
plt.legend(["Do not have heart disease", "Have heart disease"])
plt.xlabel("Chest Pain Experienced",fontsize=12)
plt.gca().set_xticklabels(
    labels=["Asymptomatic","Atypical angina","Non-anginal pain","Typical angina"]
)
plt.xticks(rotation=0)
plt.ylabel("Frequency of Disease or Not",fontsize=12)
plt.show()
```

```python
dataset.corr()
plt.figure(figsize=(8,6))
sns.heatmap(dataset.corr(), annot=True,  cmap="crest")
plt.show()
```

```python
sns.set(style="darkgrid")
sns.pairplot(dataset[['age','resting_blood_pressure','cholesterol','max_heart_rate_achieved']+['condition']], hue='condition', height=2)
plt.show()
```

```python
sns.set(style="darkgrid")
pd.crosstab(dataset.age, dataset.condition).plot(kind="bar", figsize=(20, 6), ax=plt.gca())
plt.title("Heart Disease v/s Age",fontsize=20)
plt.legend(["Do not have heart disease", "Have heart disease"])
plt.xlabel("Age",fontsize=15)
plt.ylabel("Frequency",fontsize=15)
plt.show()
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(12,6))

plt.scatter(dataset.age[dataset.condition=='Yes'], dataset.max_heart_rate_achieved[dataset.condition=='Yes'],color='salmon')
plt.scatter(dataset.age[dataset.condition=='No'], dataset.max_heart_rate_achieved[dataset.condition=='No'],color='#BAD720')

plt.title('Age v/s heart rate',fontsize=15)
plt.xlabel('Age',fontsize=12)
plt.ylabel('Max heart rate',fontsize=12)
plt.legend(['Disease','No Disease']);
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(12,6))

plt.scatter(dataset.age[dataset.condition=='Yes'], dataset.cholesterol[dataset.condition=='Yes'],color='#A569DA')
plt.scatter(dataset.age[dataset.condition=='No'], dataset.cholesterol[dataset.condition=='No'],color='#FCCC49')

plt.title('Age v/s Cholesterol',fontsize=15)
plt.xlabel('Age',fontsize=12)
plt.ylabel('Cholesterol',fontsize=12)
plt.legend(['Disease','No Disease']);
```
